### 7.4 Metacharacters
- Metacharacters are characters with a special meaning:

```python
from IPython.display import Image
Image("./0_figs/mc.png")
```

```python
# []
import re
txt = "The rain in Spain"
#Find all lower case characters alphabetically between "a" and "m":
x = re.findall("[a-m]", txt)
print(x)
```

```python
# \
import re
txt = "That will be 59 dollars"
#Find all digit characters:
x = re.findall("\d", txt)
print(x)
```

```python
# .
import re
txt = "hello world"
#Search for a sequence that starts with "he", followed by two (any) characters, and an "o":
x = re.findall("he..o", txt)
print(x)
txt = "hallowin world"
x = re.findall("ha..o", txt)
print(x)
txt = "haollowin world"
x = re.findall("ha..o", txt)
print(x)
```

```python
# ^
import re

txt = "hello world"
#Check if the string starts with 'hello':
x = re.findall("^hello", txt)
if x:
    print("Yes, the string starts with 'hello'")
else:
    print("No match")
```

```python
# $ : pattern$
import re

txt = "hello world"
#Check if the string ends with 'world':
x = re.findall("world$", txt)
if x:
    print("Yes, the string ends with 'world'")
else:
    print("No match")
```

```python
#  *: pattern*, refers to the letter just before it
import re

txt = "The rain in Spain falls mainly in the plain, plaix, plaixx!"
#Check if the string contains "ai" followed by 0 or more "x" characters:
x = re.findall("aix*", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# +: pattern+
import re

txt = "The rain in Spain falls mainly in the plain, plaix, plaixx!"
#Check if the string contains "ai" followed by 1 or more "x" characters:
x = re.findall("aix+", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# {}: pattern {times}
import re

txt = "The rain in Spain falls mainly in the plain!"
#Check if the string contains "a" followed by exactly two "l" characters:
x = re.findall("al{2}", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# |: ptrn1 | ptrn2
import re

txt = "The rain in Spain falls mainly in the plain!"
#Check if the string contains either "falls" or "stays":
x = re.findall("falls|stays", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

### 7.5 Special Sequences
- A special sequence is a \ followed by one of the characters in the list below, and has a special meaning:

```python
from IPython.display import Image
Image("./0_figs/ss.png")
```

```python
# \A: \A ptrn
import re

txt = "The rain in Spain"
#Check if the string starts with "The":
x = re.findall("\AThe", txt)
print(x)
if x:
      print("Yes, there is a match!")
else:
      print("No match")
```

```python
# \b: \bptrn
import re

txt = "The rain in Spain"
#Check if "ain" is present at the beginning of a WORD:
x = re.findall(r"\bain", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

```python
# \b: \bptrn
import re

txt = "The rain in Spain"
#Check if "ain" is present at the end of a WORD:
x = re.findall(r"ain\b", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

```python
# \B: \Bptrn
import re

txt = "The rain in Spain"
#Check if "ain" is present, but NOT at the beginning of a word:
x = re.findall(r"\Bain", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

```python
# \B: ptrn\B
import re

txt = "The rain in Spain"
#Check if "ain" is present, but NOT at the beginning of a word:
x = re.findall(r"ain\B", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

```python
# \d: "\d" 
import re

txt = "The rain in Spain"
#Check if the string contains any digits (numbers from 0-9):
x = re.findall("\d", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")

txt = "Th5e r3ai9n in S5pa0in,67"    
x = re.findall("\d", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# \D: "\D"
import re

txt = "The rain in Spain"
#Return a match at every no-digit character:
x = re.findall("\D", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

```python
# \s:  "\s"
import re

txt = "The rain in Spain"
#Return a match at every white-space character:
x = re.findall("\s", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# \w: "\w"
import re

txt = "The rain in Spain, AM56"
#Return a match at every word character (characters from a to Z, digits from 0-9, and the underscore _ character):
x = re.findall("\w", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# \W: "\W"
import re

txt = "The rain in Spain, AM56"
#Return a match at every NON word character 
 #(characters NOT between a and Z. Like "!", 
    # "?" white-space etc.):
x = re.findall("\W", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# \Z: "pattern\Z"
import re

txt = "The rain in Spain"
#Check if the string ends with "Spain":
x = re.findall("Spain\Z", txt)
print(x)
if x:
    print("Yes, there is a match!")
else:
    print("No match")
```

### 7.6  Sets
- A set is a set of characters inside a pair of square brackets [ ] with a special meaning:

```python
from IPython.display import Image
Image("./0_figs/sets.png")
```

```python
# [ptrn]
import re

txt = "The rain in Spain"
#Check if the string has any a, r, or n characters:
x = re.findall("[arn]", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# [a-z]
import re

txt = "The rain in Spain"
#Check if the string has any a, r, or n characters:
x = re.findall("[arn]", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# [^ptrn]
import re

txt = "The rain in Spain"
#Check if the string has other characters than a, r, or n:
x = re.findall("[^arn]", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# [numbers]
import re
txt = "The rain in Spain"
#Check if the string has any 0, 1, 2, or 3 digits:
x = re.findall("[0123]", txt)
print(x)
if x:
    print("Yes, there is at least one match!")
else:
    print("No match")
```

```python
# [0-9]
import re

txt = "8 times before 11:45 AM"
#Check if the string has any digits:
x = re.findall("[0-9]", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

```python
# [0-5][0-9]
import re

txt = "8 times before 11:45 AM"
#Check if the string has any two-digit numbers, from 00 to 59:
x = re.findall("[0-5][0-9]", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

```python
# [a-zA-Z]
import re

txt = "8 times before 11:45 AM"
#Check if the string has any characters from a to z lower case, and A to Z upper case:
x = re.findall("[a-zA-Z]", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

```python
# [+]
import re

txt = "8 times before+ 11:45 AM"
#Check if the string has any + characters:
x = re.findall("[+]", txt)
print(x)
if x:
      print("Yes, there is at least one match!")
else:
      print("No match")
```

#### 7.7 The findall() Function
- The **findall()** function returns a list containing all matches.
- The list contains the matches in the order they are found.
- If no matches are found, an empty list is returned:

```python
import re

txt = "The rain in Spain"
x = re.findall("ai", txt)
print(x) 
```

```python
import re
txt = "The rain in Spain"
x = re.findall("Portugal", txt)
print(x) 
```

### 7.8 The search() Function
- The search() function searches the string for a match, and returns a Match object if there is a match.
- If there is more than one match, only the first occurrence of the match will be returned
- If no matches are found, the value None is returned

```python
import re

txt = "The rain in Spain"
x = re.search("\s", txt)
print("The first white-space character"+ 
      " is located in position:", x.start())
```

```python
import re

txt = "The rain in Spain"
x = re.search("Portugal", txt)
print(x) 
```

### 7.9 The split() Function
- The split() function returns a list where the string has been split at each match
- You can control the number of occurrences by specifying the maxsplit parameter

```python
# Split at each white-space character
import re

txt = "The rain in Spain"
x = re.split("\s", txt)
print(x) 
```

```python
import re

txt = "The rain in Spain"
x = re.split("\s", txt, 1)
print(x) 
```

<div class="alert alert-block alert-warning">

### 7.10 The sub() Function
- The sub() function replaces the matches with the text of your choice
- You can control the number of replacements by specifying the count parameter:

```python
# Replace every white-space character with the number 9
import re

txt = "The rain in Spain"
x = re.sub("\s", "9", txt)
print(x) 
```

```python
# Repace with # 9 Only the first two whitespace
import re

txt = "The rain in Spain"
x = re.sub("\s", "9", txt, 2)
print(x) 
```

### 7.11 Match Object
- A Match Object is an object containing information about the search and the result.
- Note: If there is no match, the value **None** will be returned, instead of the Match Object.
- Returns a match object

- The Match object has properties and methods used to retrieve information about the search, and the result:

    - .span() returns a tuple containing the start-, and end positions of the match.
    - .string returns the string passed into the function
    - .group() returns the part of the string where there was a match

```python
import re

txt = "The rain in Spain"
x = re.search("ai", txt)
print(x) #this will print an object
```

- Example:
    - Print the position (start- and end-position) of the first match occurrence.
    - The regular expression looks for any words that starts with an upper case "S":

```python
import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.span()) 
print(x.string)
print(x.group()) 
```

```python
import re

txt = "The rain in Spain, Syria"
x = re.search(r"\bS\w+", txt)
print(x.span()) 
print(x.string)
print(x.group()) 
```

```python
import re

txt = "The rain in dSpain, fSyria"
x = re.search(r"\bS\w+", txt)
print(x)
```

- Note: If there is no match, the value None will be returned, instead of the Match Object.
    - Never use .span(), .string,or .group() if the search returns None! There will be an attribute error!
